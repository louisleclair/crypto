```python
#Using Signature verification algorithm ECDSA
def massiveFraud(p, G, pk, arr):
    res = []
    E = EllipticCurve(GF(p), [0,7])
    n = E.order()
    
    if not E.is_on_curve(pk[0], pk[1]):
        return 'fail'
    
    for m, r, s in arr:
        if not((r >= 1 and r <= n-1) or (s >= 1 and s <= n-1)):
            return 'fail'
        h_bytes = hashlib.sha256(m.encode()).digest()
        h = int.from_bytes(h_bytes, "big")
        u1 = h/s % n
        u2 = r/s % n
        point = u1*E(G) + u2*E(pk)
        x, _ = point.xy()
        if r != mod(x, n):
            res.append(m)
    return res
```

```python
def unprecedentedFraud(p, G, pk, arr, k):
    E = EllipticCurve(GF(p), [0,7])
    P = E(G)
    n = P.order()
    
    m = 'I won this election, by a lot!'
    s = [arr[i][2] for i in range(len(arr))]
    
    #First we need to recover the private key, using the fact the 2 given messages use the same k because they have the same r.
    h1 = int.from_bytes(hashlib.sha256(arr[0][0].encode()).digest(), "big")
    h2 = int.from_bytes(hashlib.sha256(arr[1][0].encode()).digest(), "big")
    delta_z = mod(h1 - h2, n)
    delta_s = mod(s[0] - s[1], n)
    k_messages = mod(delta_z/delta_s, n)
    d = mod((s[0]*k_messages - h1)/arr[0][1], n)
    
    #Now we have the private key we can forget a message using the algorithm.
    z = int.from_bytes(hashlib.sha256(m.encode()).digest(), "big")
    point = k*P
    x,_ = point.xy()
    r = mod(x, n)
    if r == 0:
        return 'fail'
    s = mod((z + Integer(r)*Integer(d))/k, n)
    if s == 0:
        return 'fail'
    return r, s
```

# Exercise 2: Cryptomancerâ€™s Tale IV: In Pursuit of Perfection

```python
pad = 20

def perfection(p, n, k, ct):
    binary_ct = ''.join([ "{:08b}".format(ord(x)) for x in base64.b64decode(ct).decode()]) 

    for divisor in divisors(n):
        #remove small divisors of n
        if divisor > 100:
            half_div = divisor / 2
            once_mod_div = k % divisor
            while half_div < once_mod_div + len(binary_ct):
                if half_div > once_mod_div:
                    half1 = int(half_div - once_mod_div) - 1
                    half2 = int(half_div - once_mod_div) + 1
                    binary_m = ''
                    while half1 >= 0 and half2 < len(binary_ct):
                        bit = '0' if binary_ct[half1] == binary_ct[half2] else '1'
                        binary_m = bit + binary_m + bit
                        half1 -= 1
                        half2 += 1
                    if binary_m[:pad] == '0'*pad:
                        binary_m_stripped = binary_m.strip('0')
                        messages = []
                        for i in range(len(binary_m_stripped)):
                            if binary_m_stripped[i:i+pad] == '0'*pad:
                                messages.append(binary_m_stripped[:i])
                                break
                                
                        for i in range(len(binary_m_stripped), 0, -1):
                            if binary_m_stripped[-pad+i:i] == '0'*pad:
                                messages.append(binary_m_stripped[i:])
                                break
                                
                        
                        for message in messages:
                            padding = (8 - (len(message) % 8))
                            for i in range(padding+1):
                                try:
                                    m_pad = i*'0' + message + '0' * (padding-i)
                                    plaintext = binascii.unhexlify('%x' % int(m_pad, 2))
                                except:
                                    pass
                                if not '\\x' in str(plaintext):
                                    return plaintext
                half_div += divisor/2
```

# Exercise 3: MBCGA

```python
# Define field
F = Integers(2)
R = PolynomialRing(F, 'y')
y = R.gen()
quot = R.quotient(y ^ 3 + y + 1, 'y')
x = quot.gen()
```

```python
def num_to_bin(num):
    """
    :param num: A number between 0 and 7.
    :return: A list with each bit for that number.
    """
    bin = BinaryStrings()
    
    return [int(c) for c in str(bin.encoding(str(num)))[-3:]]
```

```python
def str_to_bin(st, bksize=128):
    """
    :param st: A string.
    :param bksize: Block size, default 128.
    :return: A bit vector in GF(2).
    """
    bin = BinaryStrings()
    
    # Convert the plain to binary
    if st is not str:
        st = str(st)
    st_bin = str(bin.encoding(st))

    # Separate into blocks
    st_blocks = [st_bin[i:bksize + i] for i in range(0, len(st_bin), bksize)]
    while len(st_blocks[-1]) != bksize:
        print("oops")  # There is no padding anyway
        st_blocks[-1] += '0'
        
    return vector(F, st_blocks[0])  # If st more than bksize, need to update
```

```python
def bin_to_str(bn, bksize=128):
    """
    :param bn: A bit vector in GF(2).
    :param bksize: Block size, default 128.
    :return: Corresponding string.
    """
    result = ""
    for i in range(bksize / 8):
        result += chr(int(''.join([str(c) for c in bn[i * 8:(i + 1) * 8]]), 2))
    return result
```

```python
def sbox(v):
    """
    Casts the first three bits of a vector, such as the coefficients of the inversion of ax^2+bx+c mod x^3+x+1.
    :param v: A bit vector in GF(2).
    :return: A bit vector in GF(2) with the first three first bits tempered.
    """
    if v[0] == v[1] == v[2] == 0:
        return v

    a, b, c = v[0], v[1], v[2]

    pl = a * x ^ 2 + b * x + c
    pl_inv = 1/pl
    coeffs = pl_inv.list()
    c, b, a = coeffs[-3], coeffs[-2], coeffs[-1]

    return vector(F, [a, b, c] + list(v[3:]))
```

```python
def majority(v):
    """
    :param v: A 3-bit vector in GF(2).
    :return: The majority bit i.e. '0' if more than two '0', else '1'.
    """
    return v[0] if v[0] == v[1] == v[2] else 1 + v[0] + v[1] + v[2]

def sbox_map(v):
    """
    Casts the first three bits of a vector, such as the coefficients of the inversion of ax^2+bx+c mod x^3+x+1.
    :param v: A bit vector in GF(2).
    :return: A bit vector in GF(2) with the first three first bits tempered.
    """
    lhs = vector([v[0] * v[2], v[1] * v[2], v[0] * v[1]])
    maj = majority(v)
    rhs = vector([v[0] + v[1], v[0], v[0] + v[1] + v[2]])
    
    a, b, c = lhs * maj + rhs
    return vector(F, [a, b, c] + list(v[3:]))
```

```python
# Verification
for i in range(8):
    vec = num_to_bin(i)
    
    print("{} ==> {}".format(vec, sbox_map(vec)))
```

```python
def mbcga_enc(pt, af_mx, mstr_key, key_updt_mx):
    """
    McDonald's cipher.
    :param pt: A string for the plaintext.
    :param af_mx: Affine matrix.
    :param mstr_key: Master secret key
    :param key_updt_mx: Key update matrix
    :return: A string for the corresponding cipher text.
    """
    # Convert the plain to binary and define state
    state = str_to_bin(pt)

    for i in range(len(af_mx)):  # 3 times
        # S-box
        state = sbox_map(state)

        # Affine
        state =  matrix(F, af_mx[i]) * state

        # Roundkey
        key = (matrix(F, key_updt_mx) ^ (i + 1)) * vector(F, mstr_key)

        # XOR
        state = vector(F, [int(a) ^^ int(b) for (a, b) in zip(state, key)])

    return ''.join([str(b) for b in state])
```

## Questions b) and c)

```python
def mbcga_dec(ct, af_mx, pt_dum, ct_dum):
    """
    Known plain text attack on the McDonald's cipher.
    :param ct: A string for the cipher text.
    :param af_mx: Affine matrix.
    :param pt_dum: A crib.
    :param ct_dum: The crib's cipher text.
    :return: A string for corresponding the plaintext.
    """
    # Basically for each round ct + ct_dum = pt + key + pt_dum + key = pt + pt_dum
    len_af_mx = len(af_mx) - 1
    
    # Remove key
    state = vector(F, ct) + vector(F, ct_dum)
    
    # Affine
    # new_state = matrix * old_state <=> old_state = solution to AX=Y
    state = matrix(F, af_mx[len_af_mx]).solve_right(state)
    
    def inv_roundkey(state, step):
        states = []
        for i in range(2^3):
            fake_sbox_vec = vector(F, num_to_bin(i) + list(state)[3:])
            states.append(matrix(F, af_mx[step]).solve_right(fake_sbox_vec))
        return states
    
    # Inverse Roundkey
    # First step
    states = inv_roundkey(state, len_af_mx - 1)
    
    #Other inverse rounds
    for i in range(len_af_mx, 1, -1):
        states_temp = []
        for state in states:
            states_temp.append(inv_roundkey(state, i - 2))
        states = flatten(states_temp.copy())
            
    # Convert the plain to binary
    pt_dum = str_to_bin(pt_dum)
        
    # S-box
    for state in states:
        pt = sbox_map(state + sbox_map(pt_dum))
        result = bin_to_str(pt)
        try:
            if result.isalpha():
                return result
        except:
            continue
            
    return None
```
