```python
# xor of string bit
def xor(a,b):
    if a==b: 
        return '0'
    else:
        return '1'
```

```python
def decode(kmodd, half, ct_bin):
    l = len(ct_bin)
    # compute the index where there is the symmetry
    middle = half - kmodd
    first_half = int(middle) - 1
    second_half = first_half + 2
    
    m_bin = ''
    while first_half >= 0 and second_half < l:
        bit = xor(ct_bin[first_half], ct_bin[second_half])
        m_bin = bit + m_bin + bit
        first_half -= 1
        second_half += 1

    return m_bin
```

```python
import binascii
# return a string from a list of bits
def recover_string(bin_m):
    bits = bin_m.strip('0')
    
    possible_messages = []

    for i in range(len(bits), 0, -1):
        if (bits[i-40:i] == '0' * 40):
            bits_start = bits[i:]
            break
    
    possible_messages += all_possible_messages(bits_start)

    
    for i in range(len(bits)):
        if (bits[i:i+40] == '0' * 40):
            bits_end = bits[:i]
            break
            
    possible_messages += all_possible_messages(bits_end)

    for m in possible_messages:
        if not '\\x' in str(m):
            return m
        
```

```python
def all_possible_messages(bits):
    # add missing bit to have a multiple of 8 bits
    missing_bits_num = 8 - len(bits) % 8
    possible_messages = []      
    for i in range(missing_bits_num + 1):
        try:
            new_bits = '0' * i + bits + '0' * (missing_bits_num - i)
            n = int(new_bits, 2)

            possible_messages.append(binascii.unhexlify('%x' % n))
        except:
            pass
    return possible_messages
```
